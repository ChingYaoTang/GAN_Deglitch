### Baisc Idea of Autoencoders

An autoencoder is just like the picture (from [1]) where the encoder is a feature extractor to learn a latent representation from data and the decoder tries to reconstruct the data from those learned latent representations.

![Autoencoder](https://blog.keras.io/img/ae/autoencoder_schema.jpg)



###  Basic Idea of Variational Autoencoders


The objective of VAE is to learn the (true) probability distribution of data (so that we can sample from it) where data $x$ is generated from a hidden variable $z$, that is, $p(X) = \int p(X, z) dz = \int p(X|z) p(z) dz$ where $X$, $z$ are data (a random variable) and the latent representation respectively.   
  
The problem is that we have no idea the distribution of latent variable $z$. The idea of VAE is that we can infer $p(z)$ by $p(z|x)$ and use the variational inference method to approximate $p(z|x)$ by a simpler one $q(z|x)$.

Therefore, our objective is to minimize the KL-divergence (cross entropy) between two conditional probability distribution   
$D_{KL}(q(z|x)||p(z|x)) := E_q[log\ q(z|x) - log\ p(z|x)] = E_q[log\ q(z|x) - log\ p(x|z) - log\ p(z) + log\ p(x)]$.

And after rearrange the equation:  
$log\ p(x) - D_{KL}(q(z|x)||p(z|x)) = \mathcal{L}$, where $\mathcal{L} = E_q[log\ p(x|z) - log\ q(z|x)+log\ p(z)] = E_q[log\ p(x|z) - D_{KL}(q(z|x)||p(z))]$ is the lower bound of the likelihood $p(x)$ (since $D_{KL}$ is non-negative).

It is clear now the objective function (or loss function) is to maximize $\mathcal{L}$ during our training since we want to **maximize** the log-likelihood $log\ p(x)$ (MAP) plus an error term $D_{KL}(q(z|x)||p(z|x))$ due to using $q(z|x)$ to approximate $p(z|x)$. In practice, we use $-\mathcal{L}$ as our loss function and minimize it.

## Vanilla Autoencoder from [1]

Note that it is 'hard' autoencoder (i.e. not a probabilistic boundary)

```
# data preparation (MINST)


# only need image data, drop labels
(x_train, _), (x_test, _) = mnist.load_data()

# normalize to [0,1]
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.

# flatten to a vector of size 784
x_train = x_train.reshape((len(x_train), np.prod(x_train.shape[1:])))
x_test = x_test.reshape((len(x_test), np.prod(x_test.shape[1:])))


# X_train with dimension (n_instance, n_pixels)
print( x_train.shape )
print( x_test.shape )

```

```
"""
In this example, we build an autoencoder with 1-hidden layer encoder and 1-hidden layer decoder.

input -> dense(32) -> compressed rep -> dense(32) -> output

"""


from keras.layers import Input, Dense
from keras.models import Model

# this is the size of our encoded representations
encoding_dim = 32  # 32 floats -> compression of factor 24.5, assuming the input is 784 floats (MNIST data = 28*28)

# this is our input placeholder
input_img = Input(shape=(784,))
# "encoded" is the encoded representation of the input
encoded = Dense(encoding_dim, activation='relu')(input_img)
# "decoded" is the lossy reconstruction of the input
decoded = Dense(784, activation='sigmoid')(encoded)

# this model maps an input to its reconstruction
autoencoder = Model(input_img, decoded)

# this model maps an input to its encoded representation
encoder = Model(input_img, encoded)

# create a placeholder for an encoded (32-dimensional) input
encoded_input = Input(shape=(encoding_dim,))
# retrieve the last layer of the autoencoder model
decoder_layer = autoencoder.layers[-1]
# create the decoder model
decoder = Model(encoded_input, decoder_layer(encoded_input))

# use pixel-pixel binary crossentropy loss
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')

# training

tbc = TensorBoardColab()

autoencoder.fit(x_train, x_train,
                epochs=50,
                batch_size=256,
                shuffle=True,
                validation_data=(x_test, x_test),
                callbacks = [TensorBoardColabCallback(tbc)])

tbc.close()
```

```
# some visualization

def show_plots(x_test, decoded_imgs, n =10, random_select = False):
    plt.figure(figsize=(20, 4))
    
    if random_select:
        start_idx = np.random.randint(low = 0, high = len(x_test), size = 1)[0]
    else:
        start_idx = 0
    
    for i in range(n):
        # display original
        ax = plt.subplot(2, n, i + 1)
        plt.imshow(x_test[start_idx+i].reshape(28, 28))
        plt.gray()
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)

        # display reconstruction
        ax = plt.subplot(2, n, i + 1 + n)
        plt.imshow(decoded_imgs[start_idx+i].reshape(28, 28))
        plt.gray()
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
    plt.show()

```

```
# compressed representation
encoded_imgs = encoder.predict(x_test)
# reconstructed digits
decoded_imgs = decoder.predict(encoded_imgs)

show_plots(x_test = x_test, decoded_imgs = decoded_imgs, n =10)
```

## Convolutional VAE

```
# prepare data

# MNIST dataset, x_train is shape (n_samples, 28,28)
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# reshape image size to [n_samples, 28, 28, 1]
image_size = x_train.shape[1]
x_train = x_train.reshape( [-1, image_size, image_size, 1])
x_test = x_test.reshape( [-1, image_size, image_size, 1])

# normalize to [0,1]
x_train = x_train.astype('float32') / 255.
x_test = x_test.astype('float32') / 255.

# network parameters
input_shape = x_train[0].shape
kernel_size = 3
filters = 16
latent_dim = 2
epochs = 100
batch_size = 128
```

```
# reparameterization trick
# instead of sampling from Q(z|X), sample eps = N(0,I)
# z = z_mean + sqrt(var)*eps
def sampling(args):
    """Reparameterization trick by sampling fr an isotropic unit Gaussian.
    # Arguments:
        args (tensor): mean and log of variance of Q(z|X)
    # Returns:
        z (tensor): sampled latent vector
    """

    z_mean, z_log_var = args
    batch = K.shape(z_mean)[0]
    dim = K.int_shape(z_mean)[1]
    # by default, random_normal has mean=0 and std=1.0
    epsilon = K.random_normal(shape = (batch, dim))
    return z_mean + K.exp(0.5 * z_log_var) * epsilon

```

```
# construct encoder

inputs = Input(shape = input_shape, name = 'encoder_input')
x = inputs
# add two convolution + maxpool + activation layers
for i in range(2):
    # increasing filter numbers gradually
    filters *= 2
    # for same padding, p = k//2 and output size = (i+2p-k)//s +1, here k = 3, p = 1 and s = 2
    x = Conv2D(filters = filters,
               kernel_size = kernel_size, kernel_initializer = 'he_normal',
               activation = None,
               strides = (1, 1),
               padding = 'same', name = 'conv_%s'%(i+1))(x)
    # output size = (i + 2p - k)//s +1, here p = 1, k = 2 and s = 2
    x = MaxPooling2D(pool_size = (2, 2), strides = (2, 2), padding='valid', name = 'max_%s'%(i+1))(x)
    # activation layer
    x = ReLU(name = 'encoder_relu_%s'%(i+1))(x)
    

# shape info needed to build decoder model (to reverse the dimension)
shape = K.int_shape(x)
    
# generate latent vector Q(z|X)
x = Flatten(name = 'flatten_1')(x)
x = Dense(64, kernel_initializer = 'he_normal', activation='relu', name = 'fc_1')(x)
x = Dense(16, kernel_initializer = 'he_normal', activation='relu', name = 'fc_2')(x)
# mu and log_sigma, linear activation by default
z_mean = Dense(latent_dim, name='z_mean')(x)
z_log_var = Dense(latent_dim, name='z_log_var')(x)


# Sample z ~ Q(z|X)
# use reparameterization trick to push the sampling out as input
# note that "output_shape" isn't necessary with the TensorFlow backend
z = Lambda(sampling, output_shape=(latent_dim,), name='z')([z_mean, z_log_var])


# instantiate encoder model
encoder = Model(inputs, [z_mean, z_log_var, z], name='encoder')
encoder.summary()

```

```
# construct decoder

# input from latent variables (outputs of encoder)
latent_inputs = Input(shape=(latent_dim,), name='z_sampling')
# first dense layer (corresponds to the dimension of Flatten() layer in the encoder)
x = Dense(shape[1] * shape[2] * shape[3], kernel_initializer = 'he_normal', activation='relu', name = "decoder_fc")(latent_inputs)
# reshape to match the relu_2 layer in the decoder (remember we are going backwards)
x = Reshape((shape[1], shape[2], shape[3]))(x)

for i in range(2):
    # add two convolution_transpose ('deconvolution') + up_sampling + activation layers

    # output size = 2*(input_size)
    x = UpSampling2D(size = (2, 2), name = 'up_%s'%(i+1))(x)
    
    # output size = s * (i −1) + k − 2p, here p = 1, k =3, s = 1
    x = Conv2DTranspose(filters = filters,
                        kernel_size = kernel_size, kernel_initializer = 'he_normal',
                        activation = None,
                        strides = (1, 1),
                        padding = 'same', name = 'conv_t_%s'%(i+1))(x)
    # activation layer
    x = ReLU(name = 'decoder_relu_%s'%(i+1))(x)
    
    
    # reducing filter numbers gradually
    filters //= 2

decorder_outputs = Conv2DTranspose(filters = 1, # shrink filter dimensino to 1
                                   kernel_size = kernel_size, kernel_initializer = 'he_normal',
                                   activation = 'sigmoid', # recall that our inputs are normalized to [0,1]
                                   padding = 'same',
                                   name = 'decoder_output')(x)

# instantiate decoder model
decoder = Model(latent_inputs, decorder_outputs, name='decoder')
decoder.summary()


```

```

# VAE model = encoder + decoder
# instantiate VAE model
outputs = decoder(encoder(inputs)[2])
vae = Model(inputs, outputs, name='vae')
vae.summary()
```

```
def vae_loss(y_true, y_pred):
    """ 
    Define the loss function to be minimized
    loss = reconstruction loss + KL loss for each data in minibatch 
    
    Notice that 
      1. to calculate the max-likelihood we first flatten the y_true and y_pred (they are 28x28 image data) to a vector,
         then sum across the spatial dimension (all pixels) for a single instance sample
      2. for D_KL calculation, the summation is across the latent variable space (here it is only 2D since our z_mean and z_log_var are with the shape [None, 2])
    """
    # E[log P(X|z)] = - cross_entropy
    recon = K.sum(K.binary_crossentropy(K.flatten(y_true), K.flatten(y_pred)), axis = -1)
    # D_KL(Q(z|X) || P(z)); calculate in closed form as both dist. are Gaussian
    kl = 0.5 * K.sum(K.exp(z_log_var) + K.square(z_mean) - 1. - z_log_var, axis = -1)

    return recon + kl # it is actually -E[log P(X|z)] + D_KL = - L
```

```

tbc = TensorBoardColab()

vae.compile(optimizer='adam', loss = vae_loss)

vae.fit(x_train, x_train,
        epochs = epochs,
        batch_size = batch_size, 
        shuffle = True,
        validation_data = (x_test, x_test),
        callbacks = [TensorBoardColabCallback(tbc)]
       )


tbc.close()
```

```

# compressed representation
encoded_imgs = encoder.predict(x_test)
# reconstructed digits
decoded_imgs = decoder.predict(encoded_imgs[2])

show_plots(x_test = x_test, decoded_imgs = decoded_imgs, n =10)
```

```
# display a 2D manifold of the digits
def plot_manifold(models,
                 data,
                 batch_size=128,
                 model_name="vae_mnist"):
    """Plots labels and MNIST digits as function of 2-dim latent vector
    # Arguments:
        models (tuple): encoder and decoder models
        data (tuple): test data and label
        batch_size (int): prediction batch size
        model_name (string): which model is using this function
    """

    encoder, decoder = models
    x_test, y_test = data

    # display a 2D plot of the digit classes in the latent space
    z_mean, _, _ = encoder.predict(x_test,
                                   batch_size=batch_size)
    plt.figure(figsize=(15, 15))
    plt.scatter(z_mean[:, 0], z_mean[:, 1], c = y_test, cmap='tab10', alpha = 0.8)
    plt.colorbar()
    plt.xlabel("z[0]")
    plt.ylabel("z[1]")

    plt.show()


    # display a 30x30 2D manifold of digits
    n = 30
    digit_size = 28
    figure = np.zeros((digit_size * n, digit_size * n))
    # linearly spaced coordinates corresponding to the 2D plot
    # of digit classes in the latent space
    grid_x = np.linspace(-4, 4, n)
    grid_y = np.linspace(-4, 4, n)[::-1]

    # sample z_mean and z_log_var from -4 to +4
    for i, yi in enumerate(grid_y):
        for j, xi in enumerate(grid_x):
            z_sample = np.array([[xi, yi]])
            x_decoded = decoder.predict(z_sample)
            digit = x_decoded[0].reshape(digit_size, digit_size)
            figure[i * digit_size: (i + 1) * digit_size,
                   j * digit_size: (j + 1) * digit_size] = digit

    plt.figure(figsize=(20, 20))
    start_range = digit_size // 2
    end_range = n * digit_size + start_range + 1
    pixel_range = np.arange(start_range, end_range, digit_size)
    sample_range_x = np.round(grid_x, 1)
    sample_range_y = np.round(grid_y, 1)
    plt.xticks(pixel_range, sample_range_x)
    plt.yticks(pixel_range, sample_range_y)
    plt.xlabel("z[0]")
    plt.ylabel("z[1]")
    plt.imshow(figure, cmap='viridis')
    plt.show()

```

```
plot_manifold(models = (encoder, decoder),
              data = (x_test, y_test),
              batch_size=batch_size,
              model_name="vae")
```
