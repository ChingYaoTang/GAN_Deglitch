```python
%%cython --annotate 

import numpy as np
import pylab as plt

def my_random_homo_2Dsphere():
    check = 1
    
    while check:
        x = np.random.rand(3)*2-1 #產生隨機xyz座標參數
        if np.dot(x,x)<1: #取單位圓內座標
            check=0
    
    r = np.sqrt(np.dot(x,x)) #計算長度r=sqrt(x^2+y^2+z^2)
    r_xy = np.sqrt(np.dot(x[:-1],x[:-1])) #計算r*sin(theta)=sqrt(x^2+y^2)
    theta = np.arccos(x[2]/r) #因為r*cos(theta)=z, 所以theta=arccos(z/r)
    phi = np.arcsin(x[0]/r_xy) #因為r*sin(theta)*cos(phi)=x, 所以phi=arccos(x/(r*sin(theta)))
    
    return theta, phi
```

```python
theta=[]
phi=[]

for i in range(10000): #產生10k個隨機單位圓內的theta, phi座標參數
    t,p = my_random_homo_2Dsphere()
    theta.append(t)
    phi.append(p)

theta = np.array(theta)
phi = np.array(phi)

#sample於theta和phi的參數空間分佈
plt.plot(phi,theta,'b.')
plt.xlabel('phi')
plt.ylabel('theta')
plt.title('uniform distribution on 2D sphere')
#plt.savefig('random_sphere.png')
plt.show()

#球殼上的uniform分佈，在theta參數上為sin(x)/2的分佈，隨機變數變化 => Pz(z)dz=Ptheta(theta)dtheta
x = np.linspace(0,np.pi,1000)
n, bins, patches = plt.hist(theta, 50, density=False, facecolor='g', alpha=0.5)
N = len(theta)
dtheta = bins[1] - bins[0]
y = np.sin(x)/2.0*dtheta*N
plt.plot(x,y,label='sin(x)/2 distribution')
plt.xlabel('theta')
plt.ylabel('number of point')
plt.title('distribution of theta')
plt.xlim(0,np.pi)
plt.legend()
#plt.savefig('random_sphere_hist.png')
plt.show()
```
