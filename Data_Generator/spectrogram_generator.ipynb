```python
from scipy.signal import spectrogram, stft, istft, check_COLA
from scipy.misc import imsave, imread
import numpy as np
import h5py
import timeit
from sklearn import preprocessing
import matplotlib.pyplot as plt
```

```python
#time data
sample_rate = 4096
T= 64
data_length = T * sample_rate
print'length of data =', data_length, '(points)'

dt = float(1) / sample_rate
data_dir = '/Volumes/Mac/GW_glitch/' #存放產生數據的資料夾

#spectrogram data
nitv = 1./4 # proportion of 1 segments in 1 sec
NITV = (sample_rate*nitv) #length of each segment
print 'lenth of each segment =', NITV, '(points)' 

print 'the freq information contain in each freq bins =', sample_rate/NITV, '(Hz)'
print 'the time information contain in each time bins =', NITV/sample_rate, '(sec)'

novl = 1./2 # proportion of overlap in each segment
NOVL = (NITV*novl) # number of points to overlap between segments
print 'number of points to overlap between segments =', NOVL ,'(points)'
print 'time interval between time bins =', NOVL/sample_rate ,'(sec)'


lenf = sample_rate*nitv/2+1 # resolution of freq
print 'resolution of f =', lenf

lent = (data_length-NOVL)//(NITV-NOVL)+2#(nitv*T-1)/novl+1 # resolution of time
print 'resolution of t =', lent
rsp = (data_length-NOVL)%(NITV-NOVL)
print 'rest sample points =', rsp

totalpix = lenf*lent
print 'total pixel =', totalpix ,'(pixel)'
```

# batch generator

```python
def batch_generator2(mode, factor1, factor2, n, lamb):
    f = h5py.File(data_dir+'{}_{}/{}/{}_{}_waveform.h5'.format(factor1, factor2, mode, lamb, n), 'r')
    m=1
    count = True
    while count:
        data = f['data{:0>4d}'.format(m)]['TS'][...]
        
        #gw
        gw = data[0].copy()
        #glitch
        glitch = data[2].copy()
        #gw with noise & glitch
        swg = data[1]
        swg += gw
        swg += glitch
        
        yield glitch, swg, gw
        m += 1
        if m > n:
            count = False
```

```python
def batch_parameter2(mode, factor1, factor2, n, lamb):
    f = h5py.File(data_dir+'{}_{}/{}/{}_{}_waveform.h5'.format(factor1, factor2, mode, lamb, n), 'r')
    m=1
    count = True
    while count:
        para = f['data{:0>4d}'.format(m)]['parameter'][...]
        yield m, para
        m += 1
        if m > n:
            count = False
```

```python
g1 = batch_generator2('train', 1, 1, 1000, 10)
g2 = batch_parameter2('train', 1, 1, 1000, 10)
```

# spectrogram generator

```python
def specdatagenerator2(mode, factor1, factor2, n):
    #計時
    start_time = timeit.default_timer()
    
    batch = batch_generator2(mode, factor1, factor2, n) #讀取時序數據
    
    f = h5py.File(data_dir+'/{}_{}_{}_{}/spec_normalized_factor.h5'.format(mode, factor1, factor2, n),'a')
    
    cut_off_freq = 512
    len_segment = 512
    for m in range(1,n+1):
        glitch, swg, gw = next(batch)
        freqs1, bins1, spec_data1 = stft(glitch, nperseg=NITV, fs=sample_rate, noverlap=NOVL)#glitch spec
        freqs2, bins2, spec_data2 = stft(swg, nperseg=NITV, fs=sample_rate, noverlap=NOVL)#swg spec
        
        #存成jpg檔，scale會轉成[0,255]
        imsave(data_dir+'/{}_{}_{}_{}/amplitude/spec_data{:0>4d}_Am.jpg'.format(mode, factor1, factor2, n, m), np.concatenate((np.abs(spec_data1[:cut_off_freq, :len_segment]), np.abs(spec_data2[:cut_off_freq, :len_segment])), axis=1) )
        imsave(data_dir+'/{}_{}_{}_{}/phase/spec_data{:0>4d}_Ph.jpg'.format(mode, factor1, factor2, n, m), np.concatenate((np.angle(spec_data1[:cut_off_freq, :len_segment]), np.angle(spec_data2[:cut_off_freq, :len_segment])), axis=1) )
        
        #紀錄真正的scale
        glitch_scale=[]
        glitch_scale.append(np.min(np.abs(spec_data1)))
        glitch_scale.append(np.max(np.abs(spec_data1)))
        
        swg_scale=[]
        swg_scale.append(np.min(np.abs(spec_data2)))
        swg_scale.append(np.max(np.abs(spec_data2)))
        
        datam = f.create_group('data{:0>4d}'.format(m))
        datam.create_dataset('glitch_scale', data=np.array(glitch_scale))
        datam.create_dataset('swg_scale', data=np.array(swg_scale))
    
    f.close()
    stop_time = timeit.default_timer()
    print 'period(s) =', stop_time - start_time
    print('finish')
```
